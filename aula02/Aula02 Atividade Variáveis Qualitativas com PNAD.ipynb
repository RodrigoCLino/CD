```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_table("pes2013A.txt", sep=" ")
dados.head()
```

```python
n=dados.shape[0]
```

```python
#####SEXO
homens = dados[dados.V0302 == 2]
mulheres = dados[dados.V0302 == 4]

dados.V0302.value_counts()
dados.V0302 = dados.V0302.astype("category")
dados.V0302.cat.categories = ("Masculino","Feminino")

####### COR
indigena = dados[dados.V0404 == 0]
branco = dados[dados.V0404 == 2]
preto = dados[dados.V0404 == 4]
amarelo = dados[dados.V0404 == 6]
pardo = dados[dados.V0404 == 8]
sem_declaracao = dados[dados.V0404 == 9]

dados.V0404.value_counts()
dados.V0404 = dados.V0404.astype("category")
dados.V0404.cat.categories = ("indigena","branco",'preto','amarelo','pardo','sem_declaracao')

### REDE DE ENSINO
publica = dados[dados.V6002 == 2]
privada = dados[dados.V6002 == 4]
#nd = dados[dados.V6002 == ""]

dados.V6002.value_counts()
dados.V6002 = dados.V6002.astype("category")
dados.V6002.cat.categories = ("publica","privada")

### NIVEL DE ENSINO
maternal =dados[dados.V4802 == 1]
alfabetizado =dados[dados.V4802 == 2]
alfabetizado_adulto =dados[dados.V4802 == 3]
meio_EF_em8 =dados[dados.V4802 == 4]
seg_meio_EF_em8 =dados[dados.V4802 == 5]
prim_serie_9 =dados[dados.V4802 == 6]
seg_5_serie_9 =dados[dados.V4802 == 7]
final_EF_9 =dados[dados.V4802 == 8]
adulto_EF =dados[dados.V4802 == 9]
EM_preV =dados[dados.V4802 == 10]
Sup=dados[dados.V4802 == 11]
#nd=dados[dados.4802 == ]

dados.V4802.value_counts()
dados.V4802 = dados.V4802.astype("category")
dados.V4802.cat.categories = ('maternal','alfabetizado','alfabetizado_adulto','meio_EF_em8','seg_meio_EF_em8','prim_serie_9','seg_5_serie_9','final_EF_9','adulto_EF','EM_preV','Sup')

####SABE LER E ESCREVER
sim = dados[dados.V0601 == 1]
nao = dados[dados.V0601 == 3]

dados.V0601.value_counts()
dados.V0601 = dados.V0601.astype("category")
dados.V0601.cat.categories = ('sim','nao')

####Area da rede de publica

dados.V6020.value_counts()
dados.V6020 = dados.V6020.astype("category")
dados.V6020.cat.categories = ("municipal",'estadual','federal')
###UF
d = pd.DataFrame()
d['regiao']=dados.UF

d.regiao[(dados.UF >=40) &(dados.UF<50)]=1
d.regiao[(dados.UF >=30)& (dados.UF<40)]=2
d.regiao[(dados.UF >=50)& (dados.UF<60)]=3
d.regiao[(dados.UF>=20)& (dados.UF<30)]=4
d.regiao[(dados.UF >=10)& (dados.UF<20)]=5


d.regiao.value_counts()
d.regiao=d.regiao.astype('category')
d.regiao.cat.categories=('Sul','Suldeste','Centro-Oeste','Nordeste','Norte')

```

```python
# Tabelas univariadas
print ("Frequencia relativa ao sexo")
ut1 = (dados.V0302.value_counts()/n*100).round(decimals=1).reindex(["Masculino","Feminino"])
print (ut1, "\n")

print ("Frequencia relativa a cor ou raça")
ut2 = (dados.V0404.value_counts()/n*100).round(decimals=1)
print (ut2, "\n")

print ("Frequencia relativa a rede de ensino")
ut3 = (dados.V6002.value_counts()/n*100).round(decimals=1)
print (ut3, "\n")

print ("Frequencia relativa a nivel de ensino")
ut4 = (dados.V4802.value_counts()/n*100).round(decimals=1)
print (ut4, "\n")

print ("Frequencia relativa sabe ler e escrever")
ut5 = (dados.V0601.value_counts()/n*100).round(decimals=1)
print (ut5, "\n")

print ("Frequencia relativa area da rede publica")
ut6 = (dados.V6020.value_counts()/n*100).round(decimals=1)
print (ut6, "\n")

print ("Frequencia relativa a UFs")
ut7 = (d.regiao.value_counts()/n*100).round(decimals=1)
print (ut7, "\n")
```

```python
#Grafico de barras
plot= ut1.plot(kind="bar", title = "SEXO", figsize = (6,6), color = ("b", "g"))
plt.ylabel ("Freq.Relativa (em%)")
plt.xlabel ("Sexo")
```

```python
plot= ut2.plot(kind="bar", title = "COR", figsize = (6,6), color = ("b", "g","r","y","k","c",))
plt.ylabel ("Freq.Relativa (em%)")
plt.xlabel ("Cor")
```
